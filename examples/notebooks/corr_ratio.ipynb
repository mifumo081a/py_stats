```python
from py_stats import corr_ratio
```

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(np.hstack([iris.data, iris.target.reshape(-1, 1)]),
                  columns=iris.feature_names + ["label"])
df.head()
```

### 相関比をヒートマップとして出力

```python
category = "label"

cr_df, pval_df = corr_ratio(df, category)
```

```python
plt.figure()
plt.title("Correlation ratio")
g = sns.heatmap(cr_df, annot=True, vmin=0, vmax=1, center=0.5, cmap="afmhot")
g.set_xticklabels(g.get_xticklabels(), rotation=20)
g.set_yticklabels(g.get_yticklabels(), rotation=30)
plt.show()
```

```python
plt.figure()
plt.title("P-value")
g = sns.heatmap(pval_df, annot=True, vmin=0, vmax=1, center=0.5, cmap="afmhot")
g.set_xticklabels(g.get_xticklabels(), rotation=20)
g.set_yticklabels(g.get_yticklabels(), rotation=30)
plt.show()
```
